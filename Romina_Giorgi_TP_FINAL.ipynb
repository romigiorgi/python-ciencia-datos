**Ejercicio 6**\
*Realizar analisis de matriz corelacion y explicar que variable estan correlacionadas.*

```python
#Convertir la columna 'Time' a segundos
def time_to_seconds(time_str):
    parts = time_str.split(':')
    return int(parts[0]) * 3600 + int(parts[1]) * 60 + int(parts[2])

df_mens['Time'] = df_mens['Time'].apply(time_to_seconds)

# Seleccionar las columnas de interés para la matriz de correlación
columnas_interes_mens = ['Distance (KM)', 'Time']
df_seleccionado_mens = df_mens[columnas_interes_mens]

# Calcular la matriz de correlación
matriz_correlacion_mens = df_seleccionado_mens.corr()

# Imprimir la matriz de correlación
print("Matriz de correlación para hombres:")
print(matriz_correlacion_mens)
```

```python
tiempo = df_mens['Time']
distancia_km = df_mens['Distance (KM)']
plt.figure(figsize=(6, 6))
plt.scatter(distancia_km, tiempo, marker='o', color='blue', alpha=0.7)
plt.title('Gráfico de Dispersión: Tiempo vs. Distancia (KM)')
plt.xlabel('Distancia (KM)')
plt.ylabel('Tiempo (segundos)')
plt.grid(True)
plt.show()
```

```python
año_men = df_mens['Year']
plt.plot(año_men, distancia_km)
plt.title('Gráfico lineal: Distancia (KM) vs Año')
plt.xlabel('Año')
plt.ylabel('Distancia (KM)')
plt.grid(True)
plt.show()
plt.show

plt.plot(año_men, tiempo)
plt.title('Gráfico lineal: Tiempo vs Año')
plt.xlabel('Año')
plt.ylabel('Tiempo')
plt.grid(True)
plt.show()
plt.show
```

```python
import seaborn as sns
plt.figure(figsize=(4, 2))
sns.heatmap(matriz_correlacion_mens, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de Correlación')
plt.show()
```

La matriz de correlación nos muestra que a mayor distancia, menos tiempo. Pero estoy no es verdad.
Al ser un análisis historico con todos los datos, nos arroja este resultado.
Pero en realidad el deporte fue profesionalizadonse y los deportistas fueron bajando los tiempos.

Vamos a analisar el ritmo y comprobar si a lo largo del tiempo el ritmo (minutos/km) baja. Primero tengo que crear la columna de ritmo.

```python
def calcular_ritmo(tiempo, distancia):
    # Verificar que la distancia no sea cero para evitar divisiones por cero
    if distancia == 0:
        return None  # Otra opción podría ser devolver un valor predeterminado
    return tiempo / distancia

# Aplicar la función para calcular el Ritmo y agregarlo como una nueva columna 'Ritmo'
df_mens['Ritmo'] = df_mens.apply(lambda row: calcular_ritmo(row['Time'], row['Distance (KM)']), axis=1)

# Mostrar el DataFrame con la nueva columna 'Ritmo'
print(df_mens)
```

```python
df_mens['Ritmo (minutos)'] = df_mens['Ritmo'] / 60

# Mostrar el DataFrame con la nueva columna 'Ritmo (minutos)'
print(df_mens)
```

```python
# Seleccionar las columnas de interés para la matriz de correlación
columnas_interes_mens_ritmo = ['Ritmo (minutos)', 'Year']
df_seleccionado_mens_ritmo = df_mens[columnas_interes_mens_ritmo]
matriz_correlacion_mens_ritmo = df_seleccionado_mens_ritmo.corr()

# Imprimir la matriz de correlación
print("Matriz de correlación ritmo para hombres:")
print(matriz_correlacion_mens_ritmo)
```

En este caso, el coeficiente de correlación de -0.908 sugiere una fuerte relación inversa entre el ritmo y el año. A medida que corren los años, el ritmo en minutos tiende a disminuir.

```python
ritmo_mens = df_mens['Ritmo (minutos)']
plt.plot(año_men, ritmo_mens)
plt.title('Gráfico lineal: Ritmo vs Año')
plt.xlabel('Año')
plt.ylabel('Ritmo (minuto/km')
plt.grid(True)
plt.show()
plt.show
```

Conclusión:
Confirmo que a lo largo del tiempo, al profesionalizarse el deporte, bajan los ritmos de las carreras a lo largo del tiempo.
