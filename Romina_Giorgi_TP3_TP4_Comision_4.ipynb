Solicitar al usuario que ingrese la cantidad N de datos a cargar. Luego generar de manera aleatoria los siguientes datos:

* SEXO= 'MASCULINO', 'FEMENINO','NO ESPECIFICA'
* EDAD= entre 28 y 60 años
* ALTURA= entre un rango de 165.0 mts hasta 210 mts
* PESO= entre un rango de 50  y 110 kg
* HORAS DE DESCANSO diario= un número decimal

REalizar las siguientes indicaciones y extraer concluisiones sobre la información y la pooblación muestreada.

1.   Generar un array de dimensión 2 (matriz), con todos los datos ordenados, permitiendo visualizarlos como tabla.
2.   Suponiendo que la distribución de probabilidad de los datos muestrados de las N personas, se distribuyen en forma NORMAL. Encontrar la media en los datos de altura, peso, y horas de descanso, trabajo y ejercicios.
3.  Graficar la muestra o distribución de los datos  de pesos , y alturas en graficas separadas. Indicar el grado de variabilidad de sus datos
4.  Plotear de manera conjunta y encontrar si presentan correlación los datos de peso vs altura,
5. Plotear de manera conjunta los datos de horas de descanso, estudio y ejercicio



**Ayuda Como generar los datos**


```python
import numpy as np
import matplotlib.pyplot as plt

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N ,[0.4, 0.5, 0.1])
    # Por ejemplo, 40% masculino, 50% femenino, 10% no especificado)
    # Generar edades con distribución normal
    edad =np.random.normal(44, 6, N)
    edad=np.clip(np.around(edad),28,60)
    # Generar alturas, pesos y otras características
    altura = np.random.normal(180, 10, N)
    altura = np.clip(np.around(altura, decimals=1), 150.0, 210.0)

    peso = np.random.normal(80, 15, N)
    peso = np.clip(np.around(peso, decimals=1), 40.0, 130.0)

    horas_trabajo_estudio = np.random.normal(10, 2, N)
    horas_trabajo_estudio = np.clip(np.around(horas_trabajo_estudio, decimals=1), 4.0, 12.0)

    horas_ejercicio = np.random.normal(1, 0.5, N)
    horas_ejercicio = np.clip(np.around(horas_ejercicio, decimals=1), 0, 2)

    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    return datos

# Solicitar al usuario la cantidad N de datos a cargar
N = 1000

# Generar datos aleatorios para N personas con distribución normal y límites específicos
datos = generar_datos(N)

print("\n")
print("PUNTO 2")
#Obtener la Media de cada columna
#Extraer todas las columnas excepto las de género y edad
columnas_sin_genero_edad = datos [:,2:].astype(float)

#Calcular la media a lo largo del eje 0 (media de cada columna)
media_sin_genero = np.mean(columnas_sin_genero_edad, axis=0)

print("Media de todas las columnas sin género")
print("Altura  Peso  Hs descanso  Hs trabajo/estudio  Hs ejercicio")
print(media_sin_genero)

print("\n")
print("PUNTO 3")
#Graficar la muestra o distribución de los datos de pesos y alturas en graficas separadas. Indicar el grado de variabilidad de sus datos (Varianza).
peso = datos[:, 3].astype(float)
altura = datos[:, 2].astype(float)
varianza_peso = np.var(peso)
varianza_altura = np.var(altura)
print("Varianza Peso: ", varianza_peso)
print("Varianza Altura: ", varianza_altura)

plt.figure(figsize=(10, 4))  # Ajusta el tamaño de la figura
plt.subplot(1, 2, 1)  # Primera subgráfica
plt.hist(peso, bins=20, color='blue', alpha=0.7)
plt.title('Distribución de Pesos')
plt.xlabel('Peso')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)  # Segunda subgráfica
plt.hist(altura, bins=20, color='green', alpha=0.7)
plt.title('Distribución de Alturas')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

print("\n")
print("PUNTO 4")
#Plotear de manera conjunta y encontrar si presentan correlación los datos de peso vs altura.
plt.figure(figsize=(8, 6))
plt.scatter(altura, peso, color='blue', alpha=0.5)
plt.title('Diagrama de Dispersión: Peso vs Altura')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.grid(True)

correlation_coefficient = np.corrcoef(altura, peso)[0, 1] #Calcular el coeficiente de correlación
print("Coeficiente de correlación:", correlation_coefficient)
print("Coeficiente de correlación débil o nula ya que el valor es muy cercado a 0.")
plt.show()


print("\n")
print("PUNTO 5")
#Plotear de manera conjunta los datos de horas de descanso, estudio y ejercicio.
horas_descanso = datos[:, 4].astype(float)
horas_estudio = datos[:, 5].astype(float)
horas_ejercicio = datos[:, 6].astype(float)

data_to_plot = [horas_descanso, horas_estudio, horas_ejercicio] # Datos a representar en el boxplot
labels = ['Horas de descanso', 'Horas de estudio', 'Horas de ejercicio'] # Etiquetas para los ejes x
plt.boxplot(data_to_plot, labels=labels)
plt.title('Boxplot de Horas de Descanso, Estudio y Ejercicio') # Agregar título y etiquetas de ejes
plt.ylabel('Horas')
plt.show()

```
